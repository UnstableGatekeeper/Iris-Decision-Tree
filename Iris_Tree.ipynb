```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
import graphviz
```

```python
iris = load_iris()
```

```python
iris
```

```python
X, y = iris.data, iris.target
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
decision_tree = clf.fit(X, y)
```

```python
tree.plot_tree(decision_tree)
```

```python
dot_data = tree.export_graphviz(decision_tree, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True, special_characters=True)
```

```python
graph = graphviz.Source(dot_data)
```

```python
graph.render("iris")
```
